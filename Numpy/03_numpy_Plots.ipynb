Plotting NumPy Arrays

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Matplotlib

```python
#Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

```python
#scatter plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
#histogram
# it's a bar plot, it shows the distribution of contenious data. 
# Height of each plot represent the frequency of data for the specified range
# bar plot is a not contenious one, but histogram is contenious one

data = np.random.normal(100, 20, 1000)

plt.hist(data, bins=20, color='purple')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title('Histogram')
plt.show()
```

```python
#barplot
x = ['A', 'B', 'C', 'D']
y = [1, 5, 6, 2]

plt.bar(x, y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
```

```python
#Pie chart
labels = ['A', 'B', 'C', 'D']
values = [10, 50, 20, 30]

plt.pie(values, labels=labels, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart')
plt.show()
```

```python
#box plot

data = [np.random.normal(50, 10, 100), np.random.normal(80, 20, 100), np.random.normal(40, 10, 100)]

plt.boxplot(data)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box plot')
plt.show()
```

Seaborn

```python
#Line plot
import seaborn as sns
x = np.linspace(0, 10, 100)
y = np.sin(x)

sns.lineplot(x, y)
plt.show()
```

```python
#Pairplot
import pandas as pd

#Generating random data
data = np.random.randn(100, 5) 
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

sns.pairplot(df)
plt.title("Pair plot of the data")
plt.show()
```

```python
#heatmap
data = np.random.random((10, 10))

sns.heatmap(data)
plt.show()
```

```python
#Violin plot
data = np.random.normal(0, 1, 100)

sns.violinplot(data)
plt.show()
```
